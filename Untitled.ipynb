```python
#2.1
import scipy
import numpy as np
from scipy import fftpack
from scipy import signal
from scipy.io import wavfile
from matplotlib import pyplot as plt #used for plotting 
%matplotlib inline


fs = 44100
F = 500
N=500 #no of samples

#c
sound = np.zeros(fs)
n = np.arange(fs)
for i in range(1,18,2):
    if i % 3 == 0:
        sound += (1.0/i)*np.sin(2 * np.pi * n * i * F / fs + np.pi / 2)
    else:
        sound += (1.0/i)*np.sin(2 * np.pi * n * i * F / fs)
plt.plot(sound[0:N])
plt.show()
```

```python
noisy_rate = fs

y = np.arange(noisy_rate)
xlabel = y*noisy_rate/N
N2 = N/2
fft = np.fft.fft(sound[0:N]); 
fft[0] = 0 #you gotta do what you gotta do
```

```python
print(np.abs(fft[0:N2+1])/N2 > 0.2)
plt.plot(fft[0:N2])
plt.show()
```

```python
plt.stem(xlabel[0:N2+1], np.abs(fft[0:N2+1])/N2)
plt.ylabel('Magnitude')
plt.xlabel('Hz')
plt.title('FFT')
plt.show()
ifft = np.fft.ifft(fft)

plt.plot(np.real(ifft))
plt.show()
```
